## SCRAPPING OP.GG FOR THE LAST 20 MATCHES OF X PLAYER

```python
import requests
from urllib import request, response, error, parse
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
URL = "https://euw.op.gg/summoners/euw/Wanderlusted"
hdr = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
req = Request(URL,headers=hdr)
html = request.urlopen(req)
soup = BeautifulSoup(html, "html.parser")
```

```python
lst = []

for playsStats in soup.find_all("div",{"class" : ["css-ahlvdo e1iiyghw2", "css-7gbtwc e1iiyghw2"]}):
    
    registro = playsStats.find("div",{"class": "result"}).getText()
    
    equipoAzulNicks= []
    equipoRojoNicks = []
    equipoAzulCampeones= []
    equipoRojoCampeones = []
    runas = []
    jugadores= playsStats.find("div",{"class": "participants"})
    
    for runasRaw in playsStats.find('div', {"class" : "runes"}):
        
        runas.append(runasRaw.find('img',alt=True)['alt'])

    for nicks in jugadores.find_all('a', href=True):
        if len(equipoAzulNicks) < 5:
            equipoAzulNicks.append(nicks.getText())
        else:
            equipoRojoNicks.append(nicks.getText())
    
    for campeones in jugadores.find_all('img'):
        
        if len(equipoAzulCampeones) < 5:
            equipoAzulCampeones.append(campeones['alt'])
        else:
            equipoRojoCampeones.append(campeones['alt'])
    
    for kdaRaw in playsStats.find_all('div', {"class" : "kda"}):
        
        kda = kdaRaw.find("div",{"class": "k-d-a"}).getText() 
        puntuacion = kdaRaw.find("div",{"class": "ratio"}).getText()
        
    lst.append([registro, runas, kda, equipoAzulNicks, equipoRojoNicks, equipoAzulCampeones, equipoRojoCampeones])
```

```python
tablita = pd.DataFrame(lst, columns = ['Status','Runes','KDA','BlueTeamSummoners','RedTeamSummoners','BlueTeamChampions','RedTeamChampions'])

tablita
```

```python
tablita.to_csv('../data/scrapping_data.csv', sep=',', index=False)
```
