```python
%pip install riotwatcher
from riotwatcher import LolWatcher, ApiError
import pandas as pd
pd.set_option('display.max_columns', None)

import json
from pandas.io.json import json_normalize
```

```python
lol_watcher = LolWatcher('RGAPI-0c9f7fef-1ca6-4ce6-9e2c-0a7948daea5c')
my_region = 'euw1'

usuario = lol_watcher.summoner.by_name(my_region, 'Wanderlusted')
datosUsuario = pd.DataFrame.from_dict(usuario , orient='index')

datosUsuario
```

```python
my_ranked_stats = lol_watcher.league.by_summoner(my_region, usuario['id'])

type(my_ranked_stats)
```

```python
datosUsuarioRanked = pd.DataFrame (my_ranked_stats, columns = ['leagueId', 'queueType', 'tier', 'rank', 'summonerId', 'summonerName', 'leaguePoints', 'wins', 'losses', 'veteran', 'inactive', 'freshBlood', 'hotStreak', 'miniSeries'])
datosUsuarioRanked
```

```python
versions = lol_watcher.data_dragon.versions_for_region(my_region)
champions_version = versions['n']['champion']
current_champ_list = lol_watcher.data_dragon.champions(champions_version)

datosCampeones = pd.DataFrame.from_dict(current_champ_list)
datosCampeones.head(50)
```

```python
maestrias = datosCampeones.drop(['type'], axis = 1)
maestrias = maestrias.drop(['format'], axis = 1)
maestrias = maestrias.drop(['version'], axis = 1)

maestrias.head(50)
```

```python
sinkeys =[]
lst = []
for i in range(maestrias.size):
    dictio = maestrias.iloc[i].data
    dictio.pop('version')
    dictio.pop('name')
    dictio.pop('key')
    dictio.pop('blurb')
    dictio.pop('info')
    dictio.pop('image')
    dictio.pop('partype')
    dictio.pop('stats')
    lst = list(dictio.values())
    sinkeys.append(lst)
    
sinkeys
```

```python
datainapi = pd.DataFrame(sinkeys, columns = ['Name', 'Title', 'Tags'])
datainapi
```

```python
datainapi.to_csv('../data/apidata.csv', sep=',', index=False)
```
